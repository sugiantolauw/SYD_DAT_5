```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model
```

```python
titanic = pd.read_csv('C:\\Users\\richter Belmont\\Project-Zero\\SYD_DAT_5\\data\\titanic.csv', index_col='PassengerId')
```

```python
titanic
```

```python
passengers_with_age = titanic[titanic.Age.notnull()]
```

```python
age = passengers_with_age[['Age']]
```

```python
regressor = sklearn.linear_model.LogisticRegression()
```

```python
regressor.fit(age, passengers_with_age.Survived)
```

```python
plt.scatter(passengers_with_age.Age, passengers_with_age.Survived)
plt.scatter(passengers_with_age.Age, regressor.predict(age), color='red')
# it predicts everyone died
```

```python
regressor.coef_
```
